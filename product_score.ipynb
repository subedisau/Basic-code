```python
test_data['brand_response'] = ""
test_data['subcategory_response'] = ""
test_data['keywords_response'] = ""
test_data['keywords_response'] = ""
test_data['title_response'] = ""
test_data['description_response'] = ""
test_data['style_response'] = ""
test_data['category_response'] = ""
test_data['image_response'] = ""
test_data['Overall_response'] = ""
test_data['recommended_keywords'] = ""
test_data['additional_keywords'] = ""
test_data['grammer_corrected_title'] = ""
test_data['grammer_corrected_description'] = ""

for i in range(len(test_data)):
                       
    if test_data['description'][i].lower() == 'unknown' or test_data['description'][i] == '':
        test_data.loc[i,['description_response']] = 'Description missing'
        test_data.loc[i,['Overall_response']] = 'Fail'
                
        if test_data['brand'][i].lower() == 'unknown' or test_data['brand'][i] == '':           
            test_data.loc[i,['brand_response']] = 'Brand Name Missing!'
                                            
        if test_data['style'][i].lower() == 'unknown' or test_data['style'][i] == '':
            test_data.loc[i,['style_response']] = 'Style Missing!'
                    
        if test_data['title'][i].lower() == 'unknown' or test_data['title'][i] == '': 
            test_data.loc[i,['title_response']] = 'Title missing; new title cannot be recommended because the description is empty.'
            
        if test_data['title'][i].lower() != 'unknown' and test_data['title'][i] != '':
            
            # check spell and grammer
            m5 = parser.parse(test_data['title'][i])
            title = m5['result']
            if test_data['title'][i] != title:
                test_data.loc[i,['grammer_corrected_title']] = title          
                                                    
        if test_data['keywords'][i].lower() == 'unknown' or test_data['keywords'][i] == '':
            test_data.loc[i,['keywords_response']] = 'Title missing; new keywords cannot be recommended because the description is empty.'          
                                   
        if test_data['category'][i].lower() == 'unknown' or test_data['category'][i] == '':
            test_data.loc[i,['category_response']] = 'Category Missing!'
                                                        
        if test_data['subcategory'][i].lower() == 'unknown' or test_data['subcategory'][i] == '':
            test_data.loc[i,['subcategory_response']] = 'SubCategory Missing!'
                             
        if test_data['count_image'][i] < 1 or test_data['count_image'][i] == 'unknown':
            test_data.loc[i,['image_response']] = 'You may want to include atleast one image.'    
            
    if test_data['description'][i].lower() !='unknown' and test_data['description'][i]!= '':
        text = test_data['description'][i]
        q10 = nltk.sent_tokenize(text)
        list = []
        for j in q10:
            m3 = parser.parse(j)
            title = m3['result']
            list.append(title)
        list1 = " ".join(list)    
        if test_data['description'][i] != list1:            
            test_data.loc[i,['grammer_corrected_description']] = list1 
            
        if test_data['brand'][i].lower() == 'unknown' or test_data['brand'][i] == '':           
            test_data.loc[i,['brand_response']] = 'Brand Name Missing!'           
                                    
        if test_data['style'][i].lower() == 'unknown' or test_data['style'][i] == '':
            test_data.loc[i,['style_response']] = 'Style Missing!'
                                
        if test_data['title'][i].lower() == 'unknown' or test_data['title'][i] == '': 
            test_data.loc[i,['title_response']] = 'Title missing'
            test_data.loc[i,['Overall_response']] = 'Fail'
                                                                                     
        if test_data['title'][i].lower() != 'unknown' and test_data['title'][i] != '':
            # check spell and grammer
            m3 = parser.parse(test_data['title'][i])
            title = m3['result']
            if test_data['title'][i] != title:
                test_data.loc[i,['grammer_corrected_title']] = title
                       
            
        if test_data['keywords'][i].lower() == 'unknown' or test_data['keywords'][i] == '':
            
            test_data.loc[i,['keywords_response']] = 'Keywords missing'
            test_data.loc[i,['Overall_response']] = 'Fail'
            
            total_words = test_data['description'][i].split()
            total_word_length = len(total_words)
            total_sentences = tokenize.sent_tokenize(test_data['description'][i])
            total_sent_len = len(total_sentences)
            
            tf_score = {}
            for each_word in total_words:
                each_word = each_word.replace('.','')
                if each_word in tf_score:
                    tf_score[each_word] += 1
                else:
                    tf_score[each_word] = 1
            tf_score.update((x, y/int(total_word_length)) for x, y in tf_score.items())
            
            def check_sent(word,sentences):
                final = [all([w in x for w in word]) for x in sentences]
                sent_len = [sentences[i] for i in range(0, len(final)) if final[i]]
                return int(len(sent_len))
                            
            
            idf_score = {}
            for each_word in total_words:
                each_word = each_word.replace('.','')
                if each_word not in stop_words:
                    if each_word in idf_score:
                        idf_score[each_word] = check_sent(each_word, total_sentences)
                    else:
                        idf_score[each_word] = 1
                        
            idf_score.update((x, math.log(int(total_sent_len)/y)) for x, y in idf_score.items())            
            tf_idf_score = {key: tf_score[key] * idf_score.get(key, 0) for key in tf_score.keys()}
            
           # this will show the key words in dictionary form.
        
            def get_top_n(dict_elem, n):
                result = dict(sorted(dict_elem.items(), key = itemgetter(1), reverse = True)[:n]) 
                return result
            
            q = get_top_n(tf_idf_score, 50) 
            
            # this will show the keywords in list form.
            def getList(s):
                list = []
                for key in q.keys():
                    list.append(key)
                return list
            
            #get keywords list.
            m = getList(q)
            m1 = " ".join(m)
            m2 = [t for t in m1.split() if t.lower() not in stpwrd]
            m3 = ' '.join(m2)
            m4 = m3.translate(str.maketrans(punc_string, len(punc_string)*' '))
            m5 = m4.lower()
            m6 = m5.split()
            
            # get title list.
            n = test_data['title'][i].split()
            n1 = " ". join(n)
            n2 = [u for u in n1.split() if u.lower() not in stpwrd]
            n3 = " ".join(n2)
            n4 = n3.translate(str.maketrans(punc_string, len(punc_string)*' '))
            n5 = n4.lower()
            n6 = n5.split()
                        
            # Now compare the keywords list to title list.
            def intersection(lst1, lst2):
                lst3 = [value for value in lst1 if value in lst2]
                return lst3
            k = intersection(m6,n6)
                        
            # exclude the word that is in the brand.
            
            p = test_data['brand'][i].split()
            for element in p:
                if element in k:
                    k.remove(element)
            
            l = ','.join(k)
            l1 = [b for b in l.split(',') if b.lower() not in stpwrd]
            l2 = ",".join(l1)
            test_data.loc[i,['recommended_keywords']] = l2
        
        if test_data['keywords'][i].lower() != 'unknown' and test_data['keywords'][i] != '':
            
            total_words = test_data['description'][i].split()
            total_word_length = len(total_words)
            total_sentences = tokenize.sent_tokenize(test_data['description'][i])
            total_sent_len = len(total_sentences)
                          
            tf_score = {}
            for each_word in total_words:
                each_word = each_word.replace('.','')
                if each_word in tf_score:
                    tf_score[each_word] += 1
                else:
                    tf_score[each_word] = 1
            tf_score.update((x, y/int(total_word_length)) for x, y in tf_score.items()) 
            
            def check_sent(word,sentences):
                final = [all([w in x for w in word]) for x in sentences]
                sent_len = [sentences[i] for i in range(0, len(final)) if final[i]]
                return int(len(sent_len))
                       
            idf_score = {}
            for each_word in total_words:
                each_word = each_word.replace('.','')
                if each_word not in stop_words:
                    if each_word in idf_score:
                        idf_score[each_word] = check_sent(each_word, total_sentences)
                    else:
                        idf_score[each_word] = 1
            
            idf_score.update((x, math.log(int(total_sent_len)/y)) for x, y in idf_score.items())            
            tf_idf_score = {key: tf_score[key] * idf_score.get(key, 0) for key in tf_score.keys()}
            
            def get_top_n(dict_elem, n):
                result = dict(sorted(dict_elem.items(), key = itemgetter(1), reverse = True)[:n]) 
                return result
            
            q = get_top_n(tf_idf_score, 50) 
            
            def getList(s):
                list = []
                for key in q.keys():
                    list.append(key)
                return list
            
            m = getList(q)
            m1 = " ".join(m)
            m2 = [v for v in m1.split() if v.lower() not in stpwrd]
            m3 = ' '.join(m2)
            m4 = m3.translate(str.maketrans(punc_string, len(punc_string)*' '))
            m5 = m4.lower()
            m6 = m5.split()
                     
            n = test_data['title'][i].split()
            n1 = " ". join(n)
            n2 = [w for w in n1.split() if w.lower() not in stpwrd]
            n3 = " ".join(n2)
            n4 = n3.translate(str.maketrans(punc_string, len(punc_string)*' '))
            n5 = n4.lower()
            n6 = n5.split()
                        
            # compare the keywords list to title list.
            def intersection(lst1, lst2):
                lst3 = [value for value in lst1 if value in lst2]
                return lst3
            k = intersection(m6,n6)
                        
            # exclude the word that is in the brand.
            p = test_data['brand'][i].split()
            for element in p:
                if element in k:
                    k.remove(element)
                    
            r = test_data['keywords'][i].split()
            r1 = " ". join(r)
            r2 = [i for i in r1.split() if i.lower() not in stpwrd]
            r3 = " ".join(r2)
            r4 = r3.translate(str.maketrans(punc_string, len(punc_string)*' '))
            r5 = r4.lower()
            r6 = r5.split()        
                                       
            # exclude the new keywords that is in the existing keywords already. 
            
            for element1 in r6:
                if element1 in k:
                    k.remove(element1)
                       
            l = ','.join(k)
            l1 = [b for b in l.split(',') if b.lower() not in stpwrd]
            l2 = ",".join(l1)
            test_data.loc[i,['additional_keywords']] = l2
                       
        if test_data['category'][i].lower() == 'unknown' or test_data['category'][i] == '':
            test_data.loc[i,['category_response']] = 'Category Missing!'
            test_data.loc[i,['Overall_response']] = 'Fail'                    
                               
        if test_data['subcategory'][i].lower() == 'unknown' or test_data['subcategory'][i] == '':
            test_data.loc[i,['subcategory_response']] = 'SubCategory Missing!'
            test_data.loc[i,['Overall_response']] = 'Fail' 
            
        if test_data['count_image'][i] < 2 or test_data['count_image'][i] == 'unknown':
            test_data.loc[i,['image_response']] = 'You may want to include atleast one image'     
            test_data.loc[i,['Overall_response']] = 'Fail'
```
